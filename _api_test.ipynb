```python
import requests
```

```python
users = dict()
```

```python
# Create 10 users, with HRMS_ID from 2 to 11, and password as "9678781811@{HRMS_ID}"
for i in range(2, 12):
    hrms_id = str(i)
    user_password = f"{password}@{hrms_id}"
    print(f"Creating user with HRMS_ID: {hrms_id}, Password: {user_password}")
    response = requests.post(f"{testing_url}/register/", json={"HRMS_ID": hrms_id, "password": user_password})
    print(f"Response status code: {response.status_code}")
    if response.status_code == 201:
        print(f"User with HRMS_ID {hrms_id} created successfully.")
    else:
        print(f"Failed to create user with HRMS_ID {hrms_id}. Response: {response.text}")
```

```python
# Login with the created users to verify they were created successfully
for i in range(2, 12):
    hrms_id = str(i)
    user_password = f"{password}@{hrms_id}"
    print(f"Logging in with HRMS_ID: {hrms_id}, Password: {user_password}")
    response = requests.post(f"{testing_url}/login/", json={"HRMS_ID": hrms_id, "password": user_password})
    print(f"Response status code: {response.status_code}")
    if response.status_code == 200:
        print(f"User with HRMS_ID {hrms_id} logged in successfully.")
        access_token = response.json().get("access")
        refresh_token = response.json().get("refresh")
        users[hrms_id] = {"access_token": access_token, "refresh_token": refresh_token}
    else:
        print(f"Failed to log in with HRMS_ID {hrms_id}. Response: {response.text}")
```

```python
users
```

```python
# Refresh tokens for all users
for hrms_id, tokens in users.items():
    refresh_token = tokens["refresh_token"]
    print(f"Refreshing token for HRMS_ID: {hrms_id}")
    response = requests.post(f"{testing_url}/refresh/", json={"refresh": refresh_token})
    print(f"Response status code: {response.status_code}")
    if response.status_code == 200:
        new_access_token = response.json().get("access")
        users[hrms_id]["access_token"] = new_access_token
        print(f"Token refreshed successfully for HRMS_ID {hrms_id}.")
    else:
        print(f"Failed to refresh token for HRMS_ID {hrms_id}. Response: {response.text}")
```

```python
# See if we can access the protected route with the new access tokens
for hrms_id, tokens in users.items():
    access_token = tokens["access_token"]
    print(f"Accessing protected route with HRMS_ID: {hrms_id}")
    response = requests.get(f"{testing_url}/hello/", headers={"Authorization": f"Bearer {access_token}"})
    print(f"Response status code: {response.status_code}")
    if response.status_code == 200:
        print(f"Accessed protected route successfully for HRMS_ID {hrms_id}. Response: {response.json()}")
    else:
        print(f"Failed to access protected route for HRMS_ID {hrms_id}. Response: {response.text}")
```

```python
# Update the status of the first registration to "approved" and second to rejected, and alternate like this for all 10 registrations
for i in range(2, 12):
    status = "accepted" if i % 2 == 0 else "rejected"
    print(f"Updating registration {i} status to {status}")
    response = requests.post(f"{testing_url}/update_status/", headers={"Authorization": f"Bearer {access_token}"}, json={"HRMS_ID": i, "status": status})
    print(f"Response status code: {response.status_code}")
    if response.status_code != 200:
        print(f"Failed to update registration {i}. Response: {response.text}")
```

```python
# See all the registrations as admin again to verify the status updates
print(f"Accessing protected route as admin with HRMS_ID: {admin} to verify status updates")
response = requests.get(f"{testing_url}/registrations/", headers={"Authorization": f"Bearer {access_token}"})
print(f"Response status code: {response.status_code}")
if response.status_code == 200:
    print(f"Accessed protected route successfully as admin. Response: {response.json()}")
else:
    print(f"Failed to access protected route as admin. Response: {response.text}")
```

```python
# Select only the accepted registrations
accepted_users = [user for user in [user for user in response.json()] if user["user_status"] == "accepted"]
```

```python
# Create a document with the accepted users
for user in accepted_users:
    # Documents need specific fields defined in DocumentSerializer
    document_payload = {
        "document_id": f"DOCv2-{user['HRMS_ID']}",
        "name": f"New Guidelines for User {user['HRMS_ID']}",
        "version": "2.0",
        "link": "http://example.com/doc.pdf",
        "internal_link": "http://internal.example.com/doc.pdf",
        "category_names": ["General", "Onboarding"]
    }
    
    print(f"Creating document for user with HRMS_ID: {user['HRMS_ID']}")
    response = requests.post(
        f"{testing_url}/create_document/", 
        headers={"Authorization": f"Bearer {access_token}"}, 
        json=document_payload
    )
    print(f"Response status code: {response.status_code}")
    print(f"Response: {response.text}")
    if response.status_code != 201:
        print(f"Error: {response.text}")
```

```python

# We can also test fetching the documents and posts to verify they were created correctly, with all the accepted users
for user in accepted_users:
    print(f"Fetching documents for user with HRMS_ID: {user['HRMS_ID']}")
    doc_id = f"DOCv2-{user['HRMS_ID']}"
    response = requests.get(
        f"{testing_url}/documents/", 
        headers={"Authorization": f"Bearer {access_token}"},
        params={"document_ids": doc_id, "donwload": "false"}
    )
    print(f"Response status code: {response.status_code}")
    if response.status_code == 200:
        print(f"Documents fetched successfully for user {user['HRMS_ID']}. Response: {response.json()}")
    else:
        print(f"Error fetching documents for user {user['HRMS_ID']}: {response.text}")
```

```python
# Posts can also be fetched using document_id as a query parameter, to get all posts related to a document
for user in accepted_users:
    print(f"Fetching posts for user with HRMS_ID: {user['HRMS_ID']}")
    doc_id = f"DOCv2-{user['HRMS_ID']}"
    response = requests.get(
        f"{testing_url}/posts/", 
        headers={"Authorization": f"Bearer {access_token}"},
        params={"document_id": doc_id}
    )
    print(f"Response status code: {response.status_code}")
    if response.status_code == 200:
        print(f"Posts fetched successfully for user {user['HRMS_ID']}. Response: {response.json()}")
    else:
        print(f"Error fetching posts for user {user['HRMS_ID']}: {response.text}")
```
